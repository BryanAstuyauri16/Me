```python
import numpy as np
import matplotlib.pyplot as pl
import matplotlib.ticker as ticker
```

```python
x = np.linspace(-1600,1600,144444)
y = np.sin(x) * np.cos(x)
```

```python
H, W = -8, 8
H2, W2 = -7, 7

def Det_xLabel(h):
    return (-h/12) - h
H_ = Det_xLabel(H)
H_2 = Det_xLabel(H2)

def Det_yLabel(w):
    return (w/12) + w
W_ = Det_yLabel(W)
W_2 = Det_yLabel(W2)

fig, ax = pl.subplots(2,2, figsize = (24, 24), facecolor = 'Bisque',num = 10, edgecolor = 'darkgoldenrod', frameon = True, constrained_layout = True)

ax[0,0].plot(x, np.sin(x))
ax[0,0].set_xlim(H, W)
ax[0,0].set_ylim(H, W)
ax[0,0].xaxis.set_major_locator(ticker.MultipleLocator(np.pi))
ax[0,0].yaxis.set_major_locator(ticker.MultipleLocator(0.5))
ax[0,0].xaxis.set_major_formatter('{x:0.00g}')
ax[0,0].axhline(color = 'black', lw = 2.5, alpha = 0.4)
ax[0,0].axvline(color = 'black', lw = 2.5, alpha = 0.4, )
ax[0,0].text(H_, 0, 'x',fontdict ={'family': 'serif', 'color': 'k', 'weight': 'bold', 'size': '16'}, size = 16, horizontalalignment = 'center', verticalalignment='center', rotation = 'horizontal')
ax[0,0].text( 0, W_, 'y',fontdict ={'family': 'serif', 'color': 'k', 'weight': 'bold', 'size': '16'}, size = 16, horizontalalignment = 'center', verticalalignment='center', rotation = 'horizontal')
ax[0,0].spines['left'].set_position('center')
ax[0,0].spines['right'].set_color('none')
ax[0,0].spines['bottom'].set_position('center')
ax[0,0].spines['top'].set_color('none')
ax[0,0].grid()
ax[0,0].set_title('Sin(x)', fontdict ={'family': 'serif', 'color': 'darkblue', 'weight': 'bold', 'size': '16'}, loc = 'left', pad = 4)
#ax[0].secondary_xaxis('top').set_xlabel('x',fontdict ={'family': 'serif', 'color': 'k', 'weight': 'bold', 'size': '16'}, size = 16, horizontalalignment = 'center', verticalalignment='center', rotation = 'horizontal')

ax[0,1].plot(x, (1/np.tan(x)) + np.tan(x))
ax[0,1].set_xlim(H2, W2)
ax[0,1].set_ylim(H2, W2)
ax[0,1].xaxis.set_major_locator(ticker.MultipleLocator(np.pi))
ax[0,1].yaxis.set_major_locator(ticker.MultipleLocator(0.5))
ax[0,1].axhline(color = 'black', lw = 2.5, alpha = 0.4)
ax[0,1].axvline(color = 'black',lw = 2.5, alpha = 0.4)
ax[0,1].text(H_2, 0, 'x',fontdict ={'family': 'serif', 'color': 'k', 'weight': 'bold', 'size': '16'}, size = 16, horizontalalignment = 'center', verticalalignment='center', rotation = 'horizontal')
ax[0,1].text( 0, W_2, 'y',fontdict ={'family': 'serif', 'color': 'k', 'weight': 'bold', 'size': '16'}, size = 16, horizontalalignment = 'center', verticalalignment='center', rotation = 'horizontal')
ax[0,1].spines['left'].set_position('center')
ax[0,1].spines['right'].set_color('none')
ax[0,1].spines['bottom'].set_position('center')
ax[0,1].spines['top'].set_color('none')
ax[0,1].set_title('1/2*csc(2x)', fontdict ={'family': 'serif', 'color': 'darkblue', 'weight': 'bold', 'size': '16'}, loc = 'left', pad = 4)

ax[1,0].plot(x, y)
ax[1,0].set_xlim(H, W)
ax[1,0].set_ylim(H, W)
ax[1,0].xaxis.set_major_locator(ticker.MultipleLocator(np.pi))
ax[1,0].yaxis.set_major_locator(ticker.MultipleLocator(0.5))
ax[1,0].axhline(color = 'black', lw = 2.5, alpha = 0.4)
ax[1,0].axvline(color = 'black', lw = 2.5, alpha = 0.4, )
ax[1,0].text(H_, 0, 'x',fontdict ={'family': 'serif', 'color': 'k', 'weight': 'bold', 'size': '16'}, size = 16, horizontalalignment = 'center', verticalalignment='center', rotation = 'horizontal')
ax[1,0].text( 0, W_, 'y',fontdict ={'family': 'serif', 'color': 'k', 'weight': 'bold', 'size': '16'}, size = 16, horizontalalignment = 'center', verticalalignment='center', rotation = 'horizontal')
ax[1,0].spines['left'].set_position('center')
ax[1,0].spines['right'].set_color('none')
ax[1,0].spines['bottom'].set_position('center')
ax[1,0].spines['top'].set_color('none')
ax[1,0].grid()
ax[1,0].set_title('Sin(x)*Cos(x)', fontdict ={'family': 'serif', 'color': 'darkblue', 'weight': 'bold', 'size': '16'}, loc = 'left', pad = 4)

ax[1,1].plot(x, np.cos(x)+np.sin(x))
ax[1,1].set_xlim(H, W)
ax[1,1].set_ylim(H, W)
ax[1,1].xaxis.set_major_locator(ticker.MultipleLocator(np.pi))
ax[1,1].yaxis.set_major_locator(ticker.MultipleLocator(0.5))
ax[1,1].axhline(color = 'black', lw = 2.5, alpha = 0.4)
ax[1,1].axvline(color = 'black', lw = 2.5, alpha = 0.4, )
ax[1,1].text(H_, 0, 'x',fontdict ={'family': 'serif', 'color': 'k', 'weight': 'bold', 'size': '16'}, size = 16, horizontalalignment = 'center', verticalalignment='center', rotation = 'horizontal')
ax[1,1].text( 0, W_, 'y',fontdict ={'family': 'serif', 'color': 'k', 'weight': 'bold', 'size': '16'}, size = 16, horizontalalignment = 'center', verticalalignment='center', rotation = 'horizontal')
ax[1,1].spines['left'].set_position('center')
ax[1,1].spines['right'].set_color('none')
ax[1,1].spines['bottom'].set_position('center')
ax[1,1].spines['top'].set_color('none')
ax[1,1].grid()
ax[1,1].set_title('Sin(x)+Cos(x)', fontdict ={'family': 'serif', 'color': 'darkblue', 'weight': 'bold', 'size': '16'}, loc = 'left', pad = 4)



pl.show()
```

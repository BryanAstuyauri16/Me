```python
import matplotlib.pyplot as pl
from matplotlib.path import Path
import matplotlib.patches as patches
import numpy as np
import matplotlib.ticker as ticker
```

```python
def f(z, k):
    return np.dot(np.array([[ 1, k],[ 0, 1]]), z)
```

```python
z = np.array([(0, 0, 4, 4), ( 0, 2, 2, 0)])
k = 2

print(z)
print('--------------------------------------------')
print(f(z,k))
```

```python
N1 = Path([(z[0,0], z[1,0]), (z[0,1], z[1,1]), (z[0,2], z[1,2]), (z[0,3], z[1,3]), (z[0,0], z[1,0])], closed = bool)
H, W = -8.5, 8.5
H2, W2 = -8.5, 8.5

def Det_xLabel(h):
    return (-h/12) - h
H_ = Det_xLabel(H)
H_2 = Det_xLabel(H2)

def Det_yLabel(w):
    return (w/12) + w
W_ = Det_yLabel(W)
W_2 = Det_yLabel(W2)

fig, ax = pl.subplots(1,2, figsize = (16, 8), facecolor = 'Bisque',num = 10, edgecolor = 'darkgoldenrod', frameon = True, constrained_layout = True)

P = patches.PathPatch(N1, facecolor = 'white', lw=2)
ax[0].add_patch(P)
#ax[0].plot(x, -x)
ax[0].set_xlim(H, W)
ax[0].set_ylim(H, W)
#ax[0].xaxis.set_major_locator(ticker.MultipleLocator(2))
#ax[0].yaxis.set_major_locator(ticker.MultipleLocator(2))
ax[0].xaxis.set_major_formatter('{x:0.00f}')
ax[0].axhline(color = 'black', lw = 2.5, alpha = 0.4)
ax[0].axvline(color = 'black', lw = 2.5, alpha = 0.4, )
ax[0].text(H_, 0, 'x',fontdict ={'family': 'serif', 'color': 'k', 'weight': 'bold', 'size': '16'}, size = 16, horizontalalignment = 'center', verticalalignment='center', rotation = 'horizontal')
ax[0].text( 0, W_, 'y',fontdict ={'family': 'serif', 'color': 'k', 'weight': 'bold', 'size': '16'}, size = 16, horizontalalignment = 'center', verticalalignment='center', rotation = 'horizontal')
#ax[0].secondary_xaxis('top').set_xlabel('x',fontdict ={'family': 'serif', 'color': 'k', 'weight': 'bold', 'size': '16'}, size = 16, horizontalalignment = 'center', verticalalignment='center', rotation = 'horizontal')
ax[0].spines['left'].set_position('center')
ax[0].spines['right'].set_color('none')
ax[0].spines['bottom'].set_position('center')
ax[0].spines['top'].set_color('none')

verts = [(f(z,k)[0,0], f(z,k)[1,0]), (f(z,k)[0,1], f(z,k)[1,1]), (f(z,k)[0,2], f(z,k)[1,2]), (f(z,k)[0,3], f(z,k)[1,3]), (f(z,k)[0,0], f(z,k)[1,0])]
codes = [Path.MOVETO, Path.LINETO, Path.LINETO, Path.LINETO, Path.CLOSEPOLY]
N2 = Path(verts, codes)

NA = patches.PathPatch(N2, facecolor = 'white', lw=2)
ax[1].add_patch(NA)
#ax[1].plot(x, -x)
ax[1].set_xlim(H2, W2)
ax[1].set_ylim(H2, W2)
#ax[1].xaxis.set_major_locator(ticker.MultipleLocator(2))
#ax[1].yaxis.set_major_locator(ticker.MultipleLocator(2))
ax[1].axhline(color = 'black', lw = 2.5, alpha = 0.4)
ax[1].axvline(color = 'black',lw = 2.5, alpha = 0.4)
ax[1].text(H_2, 0, 'x',fontdict ={'family': 'serif', 'color': 'k', 'weight': 'bold', 'size': '16'}, size = 16, horizontalalignment = 'center', verticalalignment='center', rotation = 'horizontal')
ax[1].text( 0, W_2, 'y',fontdict ={'family': 'serif', 'color': 'k', 'weight': 'bold', 'size': '16'}, size = 16, horizontalalignment = 'center', verticalalignment='center', rotation = 'horizontal')
ax[1].spines['left'].set_position('center')
ax[1].spines['right'].set_color('none')
ax[1].spines['bottom'].set_position('center')
ax[1].spines['top'].set_color('none')

pl.show()
```
